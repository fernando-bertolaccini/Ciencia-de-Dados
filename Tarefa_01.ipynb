```python
# Importando as bibliotecas
# Imortar o pandas
import pandas as pd
# Importar o matplotlib
import matplotlib.pyplot as plt
#Mágica do matplotlib
%matplotlib inline
```

```python
# Obtendo os dados da planilha Excel
df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx', skiprows=3, skipfooter=3)
```

```python
# Verificando o início dos dados
df.head()
```

```python
# Verificando o rodapé dos dados
df.tail()
```

```python
# Verificando as colunas
df.columns
```

```python
# Renomenado as colunas para retirar os espaços
df.columns = ['Código do órgão superior', 'Nome do órgão superior', 'Código do órgão solicitante da viagem', 
              'Nome do órgão solicitante da viagem', 'N. PCDP', 'N. Reserva/Localizador', 'Data Emissão Bilhete',
              'Data Embarque', 'Valor Tarifa Comercial', 'Percentual Desconto Aplicado', 'Valor Tarifa Governo',
              'Valor Tarifa Embarque', 'Valor Bilhete', 'Companhia Aérea', 'Classe Tarifária Bilhete', 'Regra Tarifária',
              'No Show', 'Remarcado', 'Cancelado', 'Valor Multas', 'Valor Reembolso', 'Diferença de Tarifa',
              'Situação Final Bilhete']
```

# Gráficos

```python
# Qual a companhia aéria que mais vendeu passagens em out/2017
df['Companhia Aérea'].value_counts().plot.barh()
plt.xlabel('Compania Aérea')
plt.ylabel('Qtd de passagem')
plt.title('Qual a companhia que mais vendeu passagens em out/2017')
```

```python
# Gráfico de barras da quantidade de bilhetes por situação de cancelamento
df['Cancelado'].value_counts().plot.bar()
plt.title('Quantidade de bilhetes por situação de cancelamento')
```

```python
# Pizza da quantidade de bilhetes por situação
df['Situação Final Bilhete'].value_counts().plot.pie()
plt.xlabel('Situação Final Bilhete')
plt.ylabel('Qtd de bilhetes')
plt.title('Quantidade de bilhetes por situação')
```

```python
# Histograma da distribuição de frequência do valor do bilhete em R$
plt.hist(df['Valor Bilhete'], bins=12)
plt.xlabel('Valor do bilhete R$')
plt.ylabel('Frequencia absoluta')
plt.title('Histograma da distribuição de frequência di valor do bilhete em R$')
plt.show()
```
